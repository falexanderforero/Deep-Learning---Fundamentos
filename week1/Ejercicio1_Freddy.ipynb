```python
import numpy as np
def step(z):
    return 1 if z >= 0 else 0
def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y
# Valores base
w1, w2 = 1, 1
# intenta parecerse a una compuerta AND
casos = [(0,0),(0,1),(1,0),(1,1)]
bias = [-1.5, -0.5, -1.0, -2.0]

for x1, x2 in casos:
    print(f"Entrada=({x1},{x2})")
    for b in bias:
      z, y = neurona(x1, x2, w1, w2, b)
      print(f"bias={b}  z={z:.1f}  salida={y}")
```
