```python
import numpy as np
```

```python
class Perceptron:
    def __init__(self, num_features, learning_rate=0.01, epochs=1339): #constructor de la clase
        self.weights = np.zeros(num_features + 1)  # declara cero en los pesos, primer fila es el bias
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation_function(self, x):
        return 1 / (1 + np.exp(-x))   # función Sigmoide

    def sigmoid_derivative(self, output):
        return output * (1 - output) #derivada de funcion sigmoide

    def predict(self, x):
        z = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation_function(z) #aplica sigmoide al producto de entradas y pesos + bias

    def train(self, X, y): #funcion de entrenamiento

        for _ in range(self.epochs): #itera de acuerdo al numero de epocas establecido

            for xi, target in zip(X, y): #itera de acuerdo a las entradas y valores esperados de esas entradas

                #calculo del puntaje z:
                z = np.dot(xi, self.weights[1:]) + self.weights[0]
                print(f"Z: {z}")
                #calculo de la decision de salida:
                output = self.activation_function(z)
                print(f"Desicion de salida: {output}")
                #delta o error
                error = target - output
                print(f"Error: {error}")
                #funcion descenso gradiente
                gradient = error * self.sigmoid_derivative(output)
                #nuevo peso
                update = self.learning_rate * gradient

                self.weights[1:] += update * xi
                self.weights[0] += update
```

```python
# Crear un conjunto de datos simple para la demostración
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]) #Datos para el entreno
y = np.array([0, 0, 0, 1])  # AND gate -> correspondientes etiquetas de "X"

print(list(zip(X,y))) # Zip encierra los datos de entreno con su correspondiente etiqueta en tuplas individuales, 1)]
```

```python
# Inicializar y entrenar el perceptrón
perceptron = Perceptron(num_features=2)
perceptron.train(X, y)
```

```python
# Probar el modelo entrenado
for xi in X:
    pred = perceptron.predict(xi)
    print(f"Input: {xi}, Predicted: {pred:.4f}, Clasificado: {1 if pred>=0.5 else 0}")
```
